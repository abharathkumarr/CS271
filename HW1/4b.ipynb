```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_decision_boundary(w, title, ax):
    # Compute slope (a) and intercept (b)
    a = -w[1] / w[2]
    b = -w[0] / w[2]

    # Generate line points
    x1 = np.linspace(-5, 5, 100)
    x2 = a * x1 + b

    # Create grid for regions
    X1, X2 = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
    Z = np.sign(w[0] + w[1]*X1 + w[2]*X2)

    # Plot regions and boundary
    ax.contourf(X1, X2, Z, levels=[-1, 0, 1], colors=['red', 'blue'], alpha=0.2)
    ax.plot(x1, x2, 'k-', label=f'$x_2 = {a:.2f}x_1 + {b:.2f}$')
    ax.set_xlabel('$x_1$'), ax.set_ylabel('$x_2$')
    ax.set_title(title)
    ax.legend(), ax.grid(True)
```

```python
# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Case 1: w = [1, 2, 3]^T
plot_decision_boundary([1, 2, 3], '$w = [1, 2, 3]^T$', ax1)
ax1.text(0, 3, '+1', color='blue', ha='center', fontsize=12)
ax1.text(0, -3, '-1', color='red', ha='center', fontsize=12)

# Case 2: w = -[1, 2, 3]^T
plot_decision_boundary([-1, -2, -3], '$w = -[1, 2, 3]^T$', ax2)
ax2.text(0, 3, '-1', color='red', ha='center', fontsize=12)
ax2.text(0, -3, '+1', color='blue', ha='center', fontsize=12)

plt.suptitle('Perceptron Decision Boundaries', y=0.95)
plt.tight_layout()
plt.show()
```

The plots above illustrate how the decision boundary remains the same for both w = [1, 2, 3]^T and w = -[1, 2, 3]^T.
However, the classification regions for h(x) = +1 and h(x) = -1 flip when the sign of w is reversed.
