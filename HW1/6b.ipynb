```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example dataset
# Chosen example dataset: Three non-collinear points
# These points are hand-picked to be simple yet clearly demonstrate
# the difference in fit between discriminative and generative models.
# (x[0], y[0]) = (0.1, 0.5)
# (x[1], y[1]) = (1.0, 1.4)
# (x[2], y[2]) = (1.9, 1.7)
# -----------------------------------------------

x = np.array([0.1, 1.0, 1.9])
y = np.array([0.5, 1.4, 1.7])
```

```python
# Discriminative (y given x)
# This model minimizes vertical errors (in the y-direction)
# -----------------------------------------------

phi1_d, phi0_d = np.polyfit(x, y, 1)
```

```python
# Generative (x given y)
# Fit generative model (x = φ0 + φ1 * y)
# -----------------------------------------------

phi1_g, phi0_g = np.polyfit(y, x, 1)
```

```python
# Invert the generative model to make it comparable
# to the discriminative model (solve for y)
# -----------------------------------------------


phi1_g_inv = 1 / phi1_g
phi0_g_inv = -phi0_g / phi1_g
```

```python
print(f"Discriminative model: y = {phi0_d:.3f} + {phi1_d:.3f}x")
print(f"Generative model (original): x = {phi0_g:.3f} + {phi1_g:.3f}y")
print(f"Inverse Generative model: y = {phi0_g_inv:.3f} + {phi1_g_inv:.3f}x")
```

```python
# Visualization
x_line = np.linspace(0, 2, 100)
y_disc = phi0_d + phi1_d * x_line
y_gen_inv = phi0_g_inv + phi1_g_inv * x_line
```

```python
plt.scatter(x, y, color='black', label='Data points')
plt.plot(x_line, y_disc, 'r-', label='Discriminative Model')
plt.plot(x_line, y_gen_inv, 'g--', label='Generative Model (inverted)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparison: Discriminative vs. Generative Models')
plt.legend()
plt.grid(True)
plt.show()
```


**Comparison between discriminative and generative linear models fitted to the same dataset. The discriminative model (red) minimizes vertical deviations (y-errors), while the generative model (green) minimizes horizontal deviations (x-errors). The two lines differ slightly, confirming that the models are not equivalent unless the data is perfectly linear**
